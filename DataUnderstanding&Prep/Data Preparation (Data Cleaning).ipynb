#### checking that sessions are in chronological order 

```python
def convert_timestamp(df):
    
    df['time'] = (df['server_timestamp_epoch_ms'] / 1000).astype(int)
    df['time'] = pd.to_datetime(df['time'].astype(int), unit='s')
    return df
```

```python
convert_timestamp(product_df3)
```

```python
product_df3[product_df3['SessionId'] == 0]
```

```python
product_df3[product_df3['SessionId'] == 3280583]
```

```python
"""
sessions do seem to be in chrological order
"""
```

```python
#lag_features 

def lag_feature(df, group_col, target_col, offset, nan=-1):
    sr = df[target_col].shift(offset)
    sr[df[group_col] != df[group_col].shift(offset)] = nan
    return sr.fillna(nan)
```

```python
product_df3['last_event_time'] = lag_feature(product_df3, SessionId, 'time', 1, np.nan)
```
